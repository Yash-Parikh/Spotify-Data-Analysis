**Let's draw heatmap to understand relationship and impact of numerical features on each other.**

```python
plt.figure(figsize=(20,10))
sns.set(style='white')
corr = df.corr()
sns.heatmap(corr, annot=True)
```

**Heatmaps are my favourite.**

**Let's now check distributions of frequesncies of each variables between their min max range.**

```python
df.hist(figsize=(20,15))
```

**The plot below will display top 20 artists according to Avg. popularity.**

```python
plt.figure(figsize=(20,10))
sns.set(style='whitegrid')
x=df.groupby('artists')['popularity'].mean().sort_values(ascending=False).head(20)
graph_plot=sns.barplot(x, x.index)
graph_plot.set_title('Top 20 popular artists (Avg)')
graph_plot.set_ylabel('artists')
graph_plot.set_xlabel('Avg. popularity')
plt.xticks(rotation=90)
print(x)
```

**The plot below will display top 20 artists according to total popularity.**

```python
plt.figure(figsize=(20,10))
sns.set(style='whitegrid')
x=df.groupby('artists')['popularity'].sum().sort_values(ascending=False).head(20)
graph_plot=sns.barplot(x,x.index)
graph_plot.set_title('Top 20 popular artists (Sum)')
graph_plot.set_ylabel('artists')
graph_plot.set_xlabel('Total popularity')
plt.xticks(rotation=90)
print(x)
```

**Let's explore how different characteristics lost/gained their grounds in songs over past 100 years.**

```python
plt.figure(figsize=(20,10))
sns.set(style='whitegrid')
columns=['acousticness','danceability','energy','speechiness','liveness','valence']
for col in columns:
    x=df.groupby('year')[col].mean()
    graph_plot=sns.lineplot(x=x.index, y=x, label=col)
graph_plot.set_title('Music charecteristics over past 100 years')
graph_plot.set_ylabel('Meassure')
```

**From graph above, some points can be made:**

   **1) Acousticness has seen fundamental decay over the years.**
    
   **2) Danceable and energetic music is gaining its ground with every year passing by.**
    
   **3) We see so many variations in speechiness and valence in initial years and then artists seem to have kept them somewhat         constant**

```python
plt.figure(figsize=(20,10))
sns.set(style='whitegrid')
x=df.groupby('year')['loudness'].mean()
graph_plot=sns.lineplot(x=x.index, y=x, label=col)
graph_plot.set_title('Music loudness over past 100 years')
graph_plot.set_ylabel('Meassure')
```

**Let's see how popularity has varied over the years.**

```python
plt.figure(figsize=(20,10))
sns.set(style='whitegrid')
x=df.groupby('year')['popularity'].mean()
graph_plot=sns.lineplot(x=x.index, y=x, label=col)
graph_plot.set_title('Music popularity over past 100 years')
graph_plot.set_ylabel('Meassure')
```

**It is predictable and obvious that recent songs are the ones that went viral.**
