**We will predict popularity and try to quantify impacts of various features over it.**

**This implies we can let go some of our columns that has no meaning in this study.**

```python
df_copy=df.copy()
df_copy=df.drop(columns=['artists', 'id', 'mode', 'name', 'release_date'])
df_copy
```

**Let's now create dummies to convert categorical variables.**

**We will divide 100 years in to 5 parts of 20 years each**

```python
year_columns = pd.get_dummies(df_copy.year)
year_columns
```

```python
year_21_40 = year_columns.iloc[:,0:20].max(axis=1)
year_41_60 = year_columns.iloc[:,20:40].max(axis=1)
year_61_80 = year_columns.iloc[:,40:60].max(axis=1)
year_81_00 = year_columns.iloc[:,60:80].max(axis=1)
year_01_20 = year_columns.iloc[:,80:].max(axis=1)
```

**We won't include first period to let go correlation between year columns. (Same as drop first)**

```python
df_dummies=pd.concat([year_41_60,year_61_80,year_81_00,year_01_20,df_copy], axis=1)
df_dummies
```

```python
df_dummies=df_dummies.drop(columns=['year'])
```

**Let's reorder and rename our columns a bit**

```python
df_dummies.columns.values
```

```python
column_names=['year_41_60', 'year_61_80', 'year_81_00', 'year_01_20', 'acousticness', 'danceability', 'duration_ms',
       'energy','explicit', 'instrumentalness', 'key', 'liveness', 'loudness',
       'popularity', 'speechiness', 'tempo', 'valence']
```

```python
df_dummies.columns=column_names
```

```python
reordered_coulmns=['year_41_60', 'year_61_80', 'year_81_00', 'year_01_20', 'acousticness', 'danceability', 'duration_ms',
       'energy','explicit', 'instrumentalness', 'key', 'liveness', 'loudness', 'speechiness', 'tempo', 'valence', 'popularity']
```

```python
df_dummies=df_dummies[reordered_coulmns]
```
